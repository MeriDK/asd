```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
df = pd.read_excel('cars.xlsx')
df.head()
```

```python
df[df['buyer']==604]
```

```python
df1 = df[df['year'] > 2000]
pd.pivot_table(df1, values='price', index='buyer', columns='year', aggfunc=np.sum, fill_value='')
```

```python
pd.pivot_table(df1, values='price', index=['buyer'], columns=['year'], aggfunc=np.min, fill_value='')
```

```python
pd.pivot_table(df1, values='price', index=['buyer'], columns=['year'], aggfunc=np.max, fill_value='')
```

```python
pd.pivot_table(df1, values='price', index=['buyer'], columns=['year'], aggfunc=np.mean, fill_value='')
```

```python
pd.pivot_table(df1, values='price', index=['buyer'], columns=['year'], aggfunc='count', fill_value='')
```

```python
df.sort_values(by='price', ascending=False).head(10)
```

```python
df2 = df[df['year'] > 2010][['buyer', 'car']].groupby(['buyer']).count().sort_values(by='car', ascending=False).head(5)
df2
```

```python
df3 = df[df['year'] > 2010][['buyer', 'price', 'year']]
df3 = df3[df3['buyer'].isin(df2.index)]
df3.head()
```

```python
df3.pivot_table(index='buyer', columns='year', values='price', aggfunc=np.sum)
```

```python
sns.barplot(x='year', y='price', hue='buyer', data=df3, estimator=sum, ci=None)
```
